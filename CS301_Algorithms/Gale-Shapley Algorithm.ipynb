```python
def stableMatching(n, catPreferences, dogPreferences):
    # Initially, all cats and dogs are free
    freeCats = list(range(n))

    catPartner = [None] * n                     
    dogPartner = [None] * n                      
    # Each man made 0 proposals, which means that 
    # his next proposal will be to the woman number 0 in his list
    nextCatChoice = [0] * n                       
    
    # While there exists at least one free cats:
    while freeCats:
        # Pick an arbitrary free cat
        c = freeCats[0]                      
    
        cPreferences = catPreferences[c]       
        d = cPreferences[nextCatChoice[c]] 
        dPreferences = dogPreferences[d]

        currentCat = dogPartner[d]
       
        # Now "cat" tries to match with "dog". 
        # Decide whether "dog" accepts, and update the following fields
        # 1. dogPartner
        # 2. catPartner
        # 3. freeCats
        # 4. nextCatChoice
        if currentCat == None:
          #No cat case
          #"dog" accepts any proposal
          dogPartner[d] = c
          catPartner[c] = d
          #"cat" nextchoice is the next dog
          #in the catPreferences list
          nextCatChoice[c] = nextCatChoice[c] + 1
          #Delete "cat" from the 
          #free list
          freeCats.pop(0)
        else:
          #dog exists
          #Check the preferences of the 
          #current dog and that of the proposed cat's
          currentIndex = dPreferences.index(currentCat)
          hisIndex = dPreferences.index(c)
          #Accept the proposal if 
          #"cat" has higher preference in the dogPreference list
          if currentIndex > hisIndex:
             #New stable match is found for "dog"
             dogPartner[d] = c
             catPartner[c] = d
             nextCatChoice[c] = nextCatChoice[c] + 1
             freeCats.pop(0)
             freeCats.insert(0,currentCat)
          else:
             nextCatChoice[c] = nextCatChoice[c] + 1
             

    return catPartner
```
